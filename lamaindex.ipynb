```python
import time
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings

def debug_print(message):
    print(f"[DEBUG] {message}")

debug_print("Starting script execution...")

# Initialize the embedding model using HuggingFace with a timeout
try:
    debug_print("Initializing embedding model...")
    embed_model = HuggingFaceEmbedding()
    debug_print("Embedding model initialized.")
except Exception as e:
    debug_print(f"Error initializing embedding model: {e}")

# Load settings (if required, this step might be unnecessary depending on the library's API)
Settings.embed_model = embed_model

# Read and load documents from the specified directory
try:
    debug_print("Loading documents...")
    start_time = time.time()
    documents = SimpleDirectoryReader("data").load_data()
    elapsed_time = time.time() - start_time
    debug_print(f"Documents loaded in {elapsed_time:.2f} seconds.")
except Exception as e:
    debug_print(f"Error loading documents: {e}")

# Create the index from the documents using the embedding model
try:
    debug_print("Creating index from documents...")
    index = VectorStoreIndex.from_documents(documents, embed_model=embed_model)
    debug_print("Index created successfully.")
except Exception as e:
    debug_print(f"Error creating index: {e}")
```
